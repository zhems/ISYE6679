```python
import numpy as np
# Q1
A = np.array([[-2, -2,  3,  1, -2],
              [ 1,  4,  1, -1, -1]])

bT = np.array([ 0,  7])

cT = np.array([-2,  7, 14,  2, -4])

basicInd= np.array([3,4]) #x4 and x5 have indices 3 and 4
```

```python
def dual_simplex(A,bT,cT,basicInd):
    basicSize = A.shape[0] # number of constraints, m, 2
    nonbasicSize = A.shape[1] - basicSize #n-m, number of variables, 3
    n = basicSize
    m = nonbasicSize
    no_iter = 0
    nonbasicInd = np.delete(np.arange(n+m), basicInd)
    print("start")
    while True:
        no_iter += 1
        cbT = np.array(cT[basicInd]) #basic variable cost coefficients
        cnbT = np.array(cT[nonbasicInd]) #basic variable cost coefficients
        print("cbT: ",cbT)
        print("cnbT: ",cnbT)

        Ab = np.copy(A[:,basicInd]) #basis matrix
        An = np.copy(A[:,nonbasicInd]) #nonbasic variable matrix
        print("Ab:",Ab)
        print("An:",An)

        Abinv = np.linalg.inv(Ab)

        #dual variables
        pi = cbT @ Abinv
        d = (cT - pi.T @ A)[nonbasicInd]
        print("pi",pi)

        print("Checking that dual solution is feasible (d>= 0):", cT - pi.T @ A)

        #step 1&2 (select basis using indices)
        x_b = Abinv @ bT
        print ('x_b',x_b)

        #step 3
        if(all(i>=0 for i in x_b)):
            return True, basicInd

        #step 4
        l = basicInd[np.argmin(x_b)]

        print("l",l)
        #step 6
        y = (Abinv @ An).T @ A.T[l,:]
        print("y",y)
        #step 7
        if(all(i>=0 for i in y)):
            return False, None

        #step 9
        minimum_value = float('inf')
        min_index = 0
        for i in range(len(y)):
            if y[i] < 0:
                if -d[i]/y[i] <= minimum_value:
                    minimum_value , min_index = -d[i]/y[i], i
        theta_star = minimum_value
        j = nonbasicInd[min_index]
        print("j",j)
        
        basicInd = np.append(basicInd[basicInd!=l],j)
        nonbasicInd = np.append(nonbasicInd[nonbasicInd!=j],l)
        print("basicInd", basicInd, "nonbasicInd", nonbasicInd)
```

```python
dual_simplex(A,bT,cT,basicInd)
```

```python
def smart_simplex(A,bT,cT,basicInd):
    def dual_simplex(A,bT,cT,basicInd):
        basicSize = A.shape[0] # number of constraints, m, 2
        nonbasicSize = A.shape[1] - basicSize #n-m, number of variables, 3
        n = basicSize
        m = nonbasicSize
        no_iter = 0
        nonbasicInd = np.delete(np.arange(n+m), basicInd)

        while True:
            print()
            print ("start of iter "+str(no_iter)+", basis is :",basicInd)
            no_iter += 1
            cbT = np.array(cT[basicInd]) #basic variable cost coefficients
            cnbT = np.array(cT[nonbasicInd]) #basic variable cost coefficients
    #         print("cbT: ",cbT)
    #         print("cnbT: ",cnbT)

            Ab = np.copy(A[:,basicInd]) #basis matrix
            An = np.copy(A[:,nonbasicInd]) #nonbasic variable matrix
    #         print("Ab:",Ab)
    #         print("An:",An)

            Abinv = np.linalg.inv(Ab)

            #dual variables
            pi = cbT @ Abinv
            d = (cT - pi.T @ A)[nonbasicInd]
    #         print("pi",pi)

    #         print("Checking that dual solution is feasible (d>= 0):", cT - pi.T @ A)

            #step 1&2 (select basis using indices)
            x_b = Abinv @ bT
    #         print ('x_b',x_b)

            #step 3
            if(all(i>=0 for i in x_b)):
                return True, basicInd, no_iter, x_b, x_b @ cbT

            #step 4
            l = basicInd[np.argmin(x_b)]

    #         print("l",l)
            #step 6
            y = (Abinv @ An).T @ A.T[l,:]
    #         print("y",y)
            #step 7
            if(all(i>=0 for i in y)):
                return False, None, None, None, None

            #step 9
            minimum_value = float('inf')
            min_index = 0
            for i in range(len(y)):
                if y[i] < 0:
                    if -d[i]/y[i] <= minimum_value:
                        minimum_value , min_index = -d[i]/y[i], i
            theta_star = minimum_value
            j = nonbasicInd[min_index]
    #         print("j",j)
            
            basicInd = np.append(basicInd[basicInd!=l],j)
            nonbasicInd = np.append(nonbasicInd[nonbasicInd!=j],l)

    def primal_simplex(A,bT,cT,basicInd):
        
        basicSize = A.shape[0] # number of constraints, m, 2
        nonbasicSize = A.shape[1] - basicSize #n-m, number of variables, 3
        n = basicSize
        m = nonbasicSize
        no_iter = 0
        nonbasicInd = np.delete(np.arange(n+m), basicInd)
        while True:
            print()
            print ("start of iter "+str(no_iter)+", basis is :",basicInd)

            #step 1
            cbT = np.array(cT[basicInd]) #basic variable cost coefficients
            cnbT = np.array(cT[nonbasicInd]) #basic variable cost coefficients
    #         print("cbT: ",cbT)
    #         print("cnbT: ",cnbT)

            Ab = np.copy(A[:,basicInd]) #basis matrix
            An = np.copy(A[:,nonbasicInd]) #nonbasic variable matrix
            

            Abinv = np.linalg.inv(Ab)
            
            
            xhat = Abinv @ bT
    #         print ('1xhat',xhat)
            #dual variables
            pi = cbT @ Abinv
    #         print("pi",pi)

            #step 2: reduced costs
            dN = cnbT - An.T @ pi

            #step 3, current basis is optimal and can return answer
            if(all(i>=0 for i in dN)):
                xhat = Abinv @ bT
                
                return True, basicInd, no_iter, xhat, xhat @ cbT

            #step 4
            
            print('index of negative value',np.where(dN < 0))
            
            dNMinIndx=np.where(dN < 0)[0][0]
            
            j = nonbasicInd[dNMinIndx] #entering variable
    #         print ('j',j)

            #step 6
            yhat = Abinv @ A[:, j]

            #step 7: return unbounded solution
            if(all(i<=0 for i in yhat)):
                return False, None, None, None, None

            #step 9: yi = yhat, xi = Binv @ bT 
            xhat = Abinv @ bT
    #         print ('2xhat',xhat)
            theta = xhat/yhat
            
            minimum_value = float('inf')
            minIndex = 0
            for i in range(len(yhat)):
                if yhat[i]>0:
                    if xhat[i]/yhat[i]<=minimum_value:
                        minimum_value = xhat[i]/yhat[i]
                        minIndex = i
            
            theta_star = minimum_value
            l = basicInd[minIndex]
            
            basicInd = np.append(basicInd[basicInd!=l],j)

            nonbasicInd = np.append(nonbasicInd[nonbasicInd!=j],l)
            nonbasicInd=np.sort(nonbasicInd)


            #step 11,12
            Ab = np.copy(A[:,basicInd]) #basis matrix
            An = np.copy(A[:,nonbasicInd]) #nonbasic variable matrix

            #step 13


    basicSize = A.shape[0] # number of constraints, m, 2
    nonbasicSize = A.shape[1] - basicSize #n-m, number of variables, 3
    n = basicSize
    m = nonbasicSize
    no_iter = 0
    nonbasicInd = np.delete(np.arange(n+m), basicInd)
    
    cbT = np.array(cT[basicInd]) #basic variable cost coefficients
    cnbT = np.array(cT[nonbasicInd]) #basic variable cost coefficients

    Ab = np.copy(A[:,basicInd]) #basis matrix
    An = np.copy(A[:,nonbasicInd]) #nonbasic variable matrix
    
    Abinv = np.linalg.inv(Ab)

    #dual variables
    pi = cbT @ Abinv
    d = (cT - pi.T @ A)[nonbasicInd]
    if(all(i>=0 for i in d)):
        print("Using Dual Simplex")
        dual_simplex(A,bT,cT,basicInd)

    x_b = Abinv @ bT
    if(all(i>=0 for i in x_b)):
        print("Using Primal Simplex")
        primal_simplex(A,bT,cT,basicInd)
```
